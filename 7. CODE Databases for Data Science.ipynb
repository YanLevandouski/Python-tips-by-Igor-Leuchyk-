```python
import uuid
import random
```

```python
road_id = ['R' + str(i) for i in range (10)]
category = ['A'for i in range (4)] + ['B' for i in range (3)] + ['C' for i in range (3)]
random.shuffle (category)
permitted_roads = ['FD45']*4 + ['OP90']*6
random.shuffle (permitted_roads)
```

```python
table_3 = pd.DataFrame(list(zip(road_id, category, permitted_roads)), columns = ['Road_ID', 'Driving_License_Category', 'Permitted_Roads'])
table_3
```

```python
table_3.to_sql('Roads', connection, if_exists = 'replace', index = False)
connection.commit()
```

```python
select_query_1 = '''
SELECT *
FROM Vehicles
JOIN Drivers ON auto_id = Vehicle_id
WHERE Category = driver_category
'''
```

```python
df_1 = pd.read_sql_query(select_query_1, connection)
df_1
```

```python
# table above shows us correct matching between driving license category and vehicle type
```
